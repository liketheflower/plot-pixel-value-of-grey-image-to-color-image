```python
# python 3.5
# The code is modified based on the link below.
#http://stackoverflow.com/questions/32908424/how-to-draw-a-colored-pixel-on-a-grayscale-image
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
im = np.random.randint(0, 255, (40, 40))
im = im.astype(float)
#print("im",im)
I = np.dstack([im, im, im])
#np.dstack?
#print(I.shape)
#print(I)
x = 2
y = 2
I[x, y, :] = [1.0, 0.0, 0.0]
#print("new I",I)
plt.imshow(I, interpolation='nearest' )
#plt.imshow(im, interpolation='nearest', cmap='Greys')  
plt.show()
```
